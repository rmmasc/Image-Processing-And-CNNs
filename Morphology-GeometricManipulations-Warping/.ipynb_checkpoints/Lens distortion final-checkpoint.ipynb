```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
height1 = 712
width1 = 1072
center_x = height1 / 2
center_y = width1 / 2
Img1 = open('classroom.raw','rb').read()
Img1 = np.frombuffer(Img1,np.uint8)
ImageP1 = Img1[0:height1*width1]
ImageP1 = np.reshape(ImageP1,(height1,width1))
plt.figure(1,figsize=(5.5,5.5))
plt.matshow(ImageP1,cmap='gray',fignum=1)

fx = 600 
fy = 600
k1 = -0.3536
k2 = 0.1730
k3 = 0
m = 0
ImageF1 = np.zeros((height1,width1))
xd_arr = np.zeros((height1*width1,1))
yd_arr = np.zeros((height1*width1,1))
camx_arr = np.zeros((height1*width1,1)) 
camy_arr = np.zeros((height1*width1,1))
for i in range(0,height1):
    for j in range(0,width1):
        camx = (i - center_x)/fx
        camy = (j - center_y)/fy
        camx_arr [m] = camx
        camy_arr [m] = camy
        r = np.sqrt(camx**2 + camy**2)
        xd = camx*(1 + k1*(r**2) + k2*(r**4) + k3*(r**6))
        yd = camy*(1 + k1*(r**2) + k2*(r**4) + k3*(r**6))
        xd_arr [m] = xd
        yd_arr [m] = yd
        xo = xd*fx + center_x
        yo = yd*fy + center_y
        m = m+1
        ImageF1[i,j] = ImageP1[int(xo),int(yo)]

plt.figure(2,figsize=(5.5,5.5))
plt.matshow(ImageF1,cmap='gray',fignum=2)
```
